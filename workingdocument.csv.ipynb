```r
library(tidyverse)
library(repr)
library(tidymodels)
library(GGally)
library(broom)
options(repr.matrix.max.rows=6)
```

```r
telecom<-read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")|>
         select(tenure,MonthlyCharges,TotalCharges,Churn)

telecom
```

```r
monthly_charges_graph<-telecom|>
                       ggplot(aes(x=MonthlyCharges,y=TotalCharges,color=Churn))+
                       geom_point()+
                       labs(x="Monthly Charges",y="Total Charges",color="Churned or not")+
                       theme(text=element_text(size=20))

monthly_charges_graph

total_charges_graph<-telecom|>
                       ggplot(aes(x=TotalCharges,y=tenure,color=Churn))+
                       geom_point()+
                       labs(x="Total Charges",y="Tenure",color="Churned or not")+
                       theme(text=element_text(size=20))

total_charges_graph

tenure_charges_graph<-telecom|>
                       ggplot(aes(x=tenure,y=TotalCharges,color=Churn))+
                       geom_point()+
                       labs(x="tenure",y="Total Charges",color="Churned or not")+
                       theme(text=element_text(size=20))

tenure_charges_graph
             
total2_charges_graph<-telecom|>
                       ggplot(aes(x=tenure,y=TotalCharges,color=Churn))+
                       geom_point()+
                       labs(x="Tenure",y="Total Charges",color="Churned or not")+
                       theme(text=element_text(size=20))

#total2_charges_graph
```
